# Plot function for drawing

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_math_function(f, min_x, max_x, num_points = 1000):
    plt.gca().set_aspect("equal")
    f_vectorized = np.vectorize(f)
    x = np.linspace(min_x, max_x, num_points)
    y = f_vectorized(x)
    plt.plot(x, y)
    plt.show()

def plot_math_functions(functions, min_x, max_x, num_points = 1000):
    plt.gca().set_aspect("equal")
    fs_vectorized = [np.vectorize(f) for f in functions]
    x = np.linspace(min_x, max_x, num_points)
    ys = [vectorized_f(x) for vectorized_f in fs_vectorized]
    
    for y in ys:
        plt.plot(x, y)
    plt.show()
```

# Circle

```python
plot_math_functions([lambda x: np.sqrt(1 - x**2), lambda x: -np.sqrt(1 - x**2)], -1, 1, 1000)
```

```python
r = 1
phi = np.linspace(0, 2 * np.pi, 1000)
x = r * np.cos(phi)
y = r * np.sin(phi)

plt.plot(x, y)
plt.gca().set_aspect("equal")
plt.show()
```

# Trigonometric function

## Sin

```python
plot_math_function(lambda x: np.sin(x), -4, 4)
```

## Cos

```python
plot_math_function(lambda x: np.cos(x), -4, 4)
```
